# EXPLORATORY DATA ANALYSIS ON A DATASET

## 1.	Data Cleaning and Preparation:

#### A.Load the Dataset
#### B.Check for Missing Values
#### C.Data Type Validation and Correction

```python
# Step 1: Load the Dataset:
# Import necessary labriaries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset from a CSV file
df = pd.read_csv("Cardiotocographic.csv")
```

```python
# Display the first few rows
df.head()
```

```python
# Step 2: Check for Missing Values:
# Check for missing values
print("Missing Values:\n", df.isnull().sum())

# Drop missing values (if necessary)
df = df.dropna()
```

```python
# Step 3: Data Type Validation and Correction
# Check data types
print("\nData Types:\n", df.dtypes)

# Convert any object/string columns to numeric if required
df = df.apply(pd.to_numeric, errors='coerce')
```

# 2. Statistical Summary:

```python
# Summary Statistics
df.describe().T
```

# 3. Data Visualization:

#### A. Histogram & Boxplots for Distribution Analysis
#### B. Correlation Heatmap
#### C. Pairplot for Relationship Analysis

```python
# Plot histograms
df.hist(figsize=(12, 10), bins=20, edgecolor="black")
plt.suptitle("Distribution of Numerical Features", fontsize=15)
plt.show()
```

```python
# Boxplot for Outlier Detection
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.title("Boxplot of Numerical Features")
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Pairplot for Relationship Analysis
sns.pairplot(df.sample(500), diag_kind="kde")
plt.show()
```

# 4.	Pattern Recognition and Insights:

```python
# Find and display the highest correlations
correlation_matrix = df.corr()
high_correlation = correlation_matrix[abs(correlation_matrix) > 0.5]  # Adjust threshold as needed

print("Highly Correlated Features:\n", high_correlation)
```
